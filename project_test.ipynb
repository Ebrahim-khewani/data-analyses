```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv(r'/content/dataset_health/train.csv')

print(data)
```

```python
data.dropna(inplace = True)

data.isnull().sum()
```

```python
#data['Response'] = data['Response'].replace({0: 'Negative', 1: 'Positive'})

data['Response'].value_counts().plot(kind='pie', color=['red', 'green'],autopct='%1.1f%%',  startangle=90, shadow=True)
plt.title('Response Repate')
plt.xlabel('Response')
plt.ylabel('repeate')
plt.show()
```

```python
data['Driving_License'] = data['Driving_License'].replace({0: 'Dont have', 1: 'Have'})

data['Driving_License'].value_counts().plot(kind='bar' )
plt.title('Driving License')
plt.show()
```

```python
response_yes = data.loc[(data.Response==1 )]
response_yes.shape
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,4,1)
sns.countplot(x='Previously_Insured', data=response_yes)

plt.subplot(1,4,2)
sns.countplot(x='Vehicle_Damage', data=response_yes)

plt.subplot(1,4,3)
sns.countplot(x='Vehicle_Age', data=response_yes)

plt.subplot(1,4,4)
sns.countplot(x='Gender', data=response_yes)


plt.tight_layout()
plt.show()
```

```python
response_heat=response_yes[[ 'Age', 'Region_Code','Previously_Insured',  'Annual_Premium',
       'Policy_Sales_Channel', 'Vintage']]


corrMatrix = response_heat.corr()
f, ax = plt.subplots(figsize=(5, 5))
sns.heatmap(corrMatrix, annot = True)
plt.show()

```
